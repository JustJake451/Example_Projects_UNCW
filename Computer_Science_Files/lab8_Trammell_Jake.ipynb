**Problem 1:** The following cell includes an example of a line of data stored in a "Comma-separated values" format. Write a function that processes the numbers and returns the average value of all of them.

__Grades:__ 18 pts for the code (proper commenting 4 pts, function definition and usage 14 pts), and 2 pts for reflections (Total 20 pts).

```python
csv_data = '12.3,-65.87,8.9,0.0,-44.63'

# complete this function and use it
def process_csv(data):
    # define empty output list
    output_list = []
    # use a for loop to split function and append to output list
    for item in data.split(','):
        output_list.append(float(item))
    # get average of the output_list
    average_list = sum(output_list)/len(output_list)
    return average_list# update this after you complete the function

process_csv(csv_data)
```

**Problem 4:**
Assume that we are playing a simple game in which two dice are thrown and a player loses 1 dollar if the total of dice is between 2 and 7 and wins 2 dollars if it is between 8 and 12. We want to simulate this game here and see how much money a player wins or loses after playing this random dice throw game for a certain number of times. 

For this, you need to complete the __simulate_game___ function implementation in the following code cell. This function should 

- receive an integer number as input for simulating the game for certain number of times (e.g. simulate_game(10) means simulate this game 10 times)
- use random functions to generate two numbers between 1 and 6 (two dice) for each round of the game 
- implement the win/lose logic of the game and keep track of the total money won or lost in the game and return it as output

Once you the code is working, you should see that you are making money sometimes and losing money sometimes randomly.

__Just for fun:__ Let's see what happens if we run the game for a very long time (1 million times for example). See what are the odds of gaining and losing money in this condition in the long run (it is called expected value)!

__Grades:__ 22.5 pts for the code (proper commenting 5 pts, function definition 17.5 pts), and 2.5 pts for reflections (Total 25 pts).


```python
# rolling of dice.
import random
# Constants for the minimum and maximum random numbers
MIN = 1
MAX = 6
def main():
    total_games = 10
    # call the simulate_game function here and put the results in my_money
    my_money = simulate_game(total_games)
    print('after playing the game,',total_games,' times I got',my_money, 'dollars which is on average', my_money/total_games,'dollars per game')

def simulate_game(num_of_simulations): 
    # set num_of_simulations equal to a count
    rounds_ = num_of_simulations  
    # keep track of total money earned
    money_earned = 0
    # create a while loop that counts down number of rounds
    while rounds_ > 0:
        # roll the dice
        first_dice = random.randint(MIN, MAX)
        second_dice = random.randint(MIN, MAX)
        dice_total = first_dice + second_dice
        # use if statements for win/loss logic
        if dice_total >= 2 and dice_total <= 7:
            money_earned -= 1 
        elif dice_total >= 8 and dice_total <= 12:
            money_earned += 2
        # remove one round per loop
        rounds_ -= 1 
    return  money_earned # returns money earned 
    
# Call the main function.
main()
```

**Problem 5:**
Use the __math module__ functions to complete the remaining calculations in the calculate_x_y() function to plot a spiral!


You will need to implement:

# $r = {\rm e}^{-0.1\theta}$

# $x = r\cos(\theta)$

# $y = r\sin(\theta)$

And similar to the previous labs, make a list of $x$ and $y$ values to see the plot.


__Grades:__ 13.5 pts for the code (proper commenting 2 pts, each math equation 3.5 pts, updating the lists correctly 1 pt), and 1.5 pts for reflections (Total 15 pts).

```python
#import the matplotlib module for plotting the results
import matplotlib.pyplot as plt
import math

def main():
    # calculate two lists for x and y 
    x, y = calculate_x_y()

    # plot the results using the obtained lists
    plot_results(x,y)


# calculating the x and y values
def calculate_x_y():
    x_list = [] # empty list for x values
    y_list = [] # empty list for y values
    # complete the calculations for x_list and y_list here via a for loop
    # loop through some angular positions for theta
    total_segments = 1000
    for segment in range(total_segments):
        theta = 0.02*math.pi*segment
        ##############################################
        # TODO: complete the math implementation here
        # calculate r
        r = math.exp(theta * -0.1)
        # calculate x
        x = r*math.cos(theta)
        # calculate y 
        y = r*math.sin(theta)
        # append x and y to their respective lists
        x_list.append(x)
        y_list.append(y) 
       ###############################################
    # return the lists
    return x_list, y_list

# plotting the results
def plot_results(x_list, y_list):
    plt.plot(x_list,y_list)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
main()
```
