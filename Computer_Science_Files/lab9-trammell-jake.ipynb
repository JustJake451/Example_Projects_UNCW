**Problem 1:**
Complete the read_x_y_files function to read a set of x and y coordinates from p1_x.txt and p1_y.txt files and return two lists containing the x and y coordinates. You will see a spiral in the plot if you read and return the values correctly.


__Grades:__ 18 pts for the code (proper comments 4 pts, opening, reading, processing, and closing the files correctly 14 pts), and 2 pts for reflections (Total 20 pts).


```python
import matplotlib.pyplot as plt

def main():
    # read the x and y values from the files
    x_file = '/kaggle/input/lab-9-dataset/p1_x.txt'# complete the file name and address based on your folder name
    y_file = '/kaggle/input/lab-9-dataset/p1_y.txt'# complete the file name and address based on your folder name
    x, y = read_x_y_files(x_file, y_file)
    
    # plot the results using the obtained lists
    plot_results(x,y)
    
    
# reading the x and y values
def read_x_y_files(x_file, y_file):
    # Complete this function
    # Set up an empty list
    x_list = []
    y_list = []
    # Open the files
    x_values = open(x_file, 'r')
    y_values = open(y_file, 'r')
    # Convert string to float and append to x_list
    for line in x_values:
    # convert to float
        num1 = float(line)
        x_list.append(num1)
    # Convert string to float and append to y_list
    for line in y_values:
        #convert to float
        num2 = float(line)
        y_list.append(num2)
    # Close the files
    x_values.close()
    y_values.close()
    # Return the values
    return x_list, y_list

# plotting the results
def plot_results(x_list, y_list):
    plt.plot(x_list,y_list)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
main()
```

**Problem 5:** In this problem, you need to process the p5.txt file according to the following descriptions and instructions.

This file contains the information for 100 files with different extensions (e.g. ipynb, doc, txt similar to the following example)




In each line, a file name and a file size in kilobytes are included and the entries are separated by a tab.  

You need to process the file and determine the total size of each type of file (files with certain extensions). A list is provided in the following cell that includes all the file extension. You can use it your solution. 

__Hint:__ Use the "endswith" function for strings to check each files extension. See the following examples if needed


https://www.geeksforgeeks.org/python-string-endswith-method/


__Grades:__ 22.5 pts for the code (proper commenting 4 pts, opening the file in the correct mode, processing, and closing the files correctly 18.5 pts), and 2.5 pts for reflections (Total 25 pts).

```python
#extensions = ['.txt', '.py', '.ipynb', '.doc', '.jpg']

# Define the main function
def main():
    file_name = '/kaggle/input/lab-9-dataset/p5.txt'
    extensions = ['.txt', '.py', '.ipynb', '.doc', '.jpg']
    calc_extension_size(file_name, extensions)

# Define calc_extension_size function
def calc_extension_size(file_name, extensions):
    # Initialize counters for each file type
    txt_total = 0
    py_total = 0
    ipynb_total = 0
    doc_total = 0
    jpg_total = 0 
    # open the file and save to a variable
    infile = open(file_name, 'r')
    # Create a for loop to read through each line 
    for line in infile:
        parts = line.strip().split('\t') #splits into two parts by tab
        if len(parts) != 2: # skips line w/o two parts
            continue
        # Extract filename and size from the parts    
        filename = parts[0]
        size = int(parts[1]) # converts to integer
        # Add size to total based on extension
        if filename.endswith('.txt'):
            txt_total += size
        elif filename.endswith('.py'):
            py_total += size
        elif filename.endswith('.ipynb'):
            ipynb_total += size
        elif filename.endswith('.doc'):
            doc_total += size
        elif filename.endswith('.jpg'):
            jpg_total += size
    # Close the file
    infile.close()
    # Print the results
    print('Total size of .txt files:', txt_total, 'KB')
    print('Total size of .py files:', py_total, 'KB')
    print('Total size of .ipynb files:', ipynb_total, 'KB')
    print('Total size of .doc files:', doc_total, 'KB')
    print('Total size of .jpg files:', jpg_total, 'KB')

# Call the main function
main()
```

```python
# Define a function
def multiplication_table():
    # open the new file in working and save as variabble
    outfile = open('/kaggle/working/multiplication_table.txt', 'w')
    # Create a for loop for multiplication table
    for y in range(1,11): # rows
        for x in range(1,11): # columns
            # Multiply x by y then convert to a string
            outfile.write(str(y*x) + "\t")
        #Add a line after each row
        outfile.write("\n")
    # Close the file
    outfile.close()
    # Print message to test if code runs
    print("The multiplication table has been saved!")

# Calls the function
multiplication_table()
```
