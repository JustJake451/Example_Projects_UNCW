**Problem 2:**
For this problem, you need to write a program that reads and processes the p2.txt file (containing a list of majors for 50 students). Your program must 

- make a __dictionary__ that includes student majors as keys and the number of enrolled students in those major as values.
- make a __set__ of the unique majors listed in the file.
- add a new 'ISE' major to the set if it is not already included in the file.


__Grades:__ 27 pts for the code (proper comments 5 pts, opening, reading, processing, and closing the files correctly 10 pts, correctly creating the new dictionary and set 12 pts), and 3 pts for reflections (Total 30 pts).



```python
def main():
    # Complete your code here
    # open the file and save to a variable
    infile = open('/kaggle/input/lab-9-dataset/p2.txt', 'r')
    # create a blank dictionary
    majors_list = {}
    # create an empty set
    majors_unique = set()
    # read the data line by line using a for loop
    for line in infile:
        # remove the newline after each entry
        major = line.rstrip('\n')
        # Add to the dictionary or update count
        if major in majors_list:
            majors_list[major] += 1 # adds to the dictionary 
        else: 
            majors_list[major] = 1 # sets the major count at 1
        # update the set 
        majors_unique.add(major)
    # close the file
    infile.close()

    # add the ISE Major
    majors_unique.add('ISE')

    # print the set to check
    print(majors_unique)
main()
```

__Step 1:__ Use the two lists in the following code cell to create a dictionary. This dictionary includes the UNCW chancellors as keys and the years that they led the university as values. __USE LOOPS (You CANNOT copy and paste the list values and manually create a dictionary)__.

__Step 2:__ Use the dictionary created in step 1 to look up some information about the UNCW chancellors. The additional code that you add needs to 

- ask the user for the last name of one of the UNCW chancellors (current or former)
- shows the time period during which the specific chancellor led UNCW and if the name is not already in the chancellor_names list, shows this message: 'This person has never been a UNCW chancellor!'.


__Grades:__ 27 pts for the code (proper comments 6 pts, reading and processing the inputs via exception handling methods correctly 21 pts), and 3 pts for reflections (Total 30 pts).

```python
def main():
    
    chancellor_names = ['Hamilton', 'Hoggard', 'Randall', 'Wagoner', 'Leutze', 'DePaolo', 'Miller', 'Sederburg', 'Sartarelli', 'Volety']
    chancellor_years = ['1947-49','1949-58', '1958-68',  '1968-90', '1990-2003', '2003-11', '2011-14', '2014-15', '2015-2022', '2022-pres.']
    # complete your code here
    # create a blank dictionary
    uncw_chancellors = {}
    # use a for loop to fill the dictionary
    for x in range(len(chancellor_names)):
        # pair each chancellor with their corresponding values
        uncw_chancellors[chancellor_names[x]] = chancellor_years[x]
    # ask user for chancellors last name
    user_input = input('Enter a last name for a UNCW chancellor:')
    # check for user_input in the new dictionary and return message
    if user_input in uncw_chancellors:
        print(user_input, 'led UNCW during', uncw_chancellors[user_input])
    else:
        print('This person has never been a UNCW chancellor')
main()
```

**Problem 4:**

Use the three sets in the following code cell to __perform some set operations__. Your program needs to find and display

1. Junk foods: foods that contain high levels of both fat and carbs.
2. Foods suitable for keto diets: foods that contain high fat but not high carbs
3. Foods suitable for low-carb diets: foods that contain high fat or protein but not high carbs
4. Foods suitable for intense training which are not including junk foods: foods either containing high protein or high carbs but excluding the junk foods.


__Grades:__ 18 pts for the code (proper comments 4 pts, correct solution for each of the four required set 3.5 pts), and 2 pts for reflections (Total 20 pts).

```python
def main():
    # DO NOT CHANGE THIS PART
    high_carb = set(['Bread', 'Pizza', 'Beans', 'Banana', 'Rice', 'Fries'])
    high_fat = set(['Nuts', 'Salmon', 'Whole eggs', 'Pizza', 'Steak', 'Fries'])
    high_protein = set(['Steak','Salmon', 'Whole eggs','Chicken', 'Whey', 'Beans', 'Nuts'])
    # Complete your set operations here
    
    # create a set of junk foods: high in fat and carbs
    junk_food = high_carb.intersection(high_fat)
    # create a set for keto diet: high fat but low carbs
    keto_food = high_fat.difference(high_carb)
    # create a set for low-carb diet: high fat/protein but not high carbs
    low_carb_food = (high_fat.union(high_protein)).difference(high_carb)
    # create set for intense training: high protein/crabs but no junk food
    intensive_training_food = (high_protein.union(high_carb)).difference(junk_food)
    # print the final results
    print('Junk-foods:', junk_food)
    print('Keto Diet Foods:', keto_food)
    print('Low-carb Diet:', low_carb_food)
    print('Intensive Training Foods:', intensive_training_food)
main()
```
