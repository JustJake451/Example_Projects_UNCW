**Problem 1:** Scientists measure an object’s mass in kilograms and its weight in newtons. If you know
the amount of mass of an object in kilograms, you can calculate its weight in newtons with
the following formula:

weight = mass x 9.8

Write a program that asks the user to enter an object’s mass, then calculates its weight. If
the object weighs more than 500 newtons, display a message indicating that it is too heavy.
If the object weighs less than 100 newtons, display a message indicating that it is too light.

__Grades:__ 9 pts for the code and comments and 1 pt for reflections (Total 10 pts)

```python
# Stores the constant for weight conversion formula
CONSTANT = 9.8
# Reads input from user
mass = float(input('Enter the object\'s mass in kilograms (kg): '))
# Converts kilograms to newtons
weight = mass * CONSTANT
# Compares the weight of object to newton amounts listed
if weight > 500:
    print('The object is too heavy!')
elif weight < 100:
    print('The object is too light!')
else:
    print("The object weighs:",format(weight, '.2f'), "newtons.")
```

Instead of using the else function to display the weight in newtons, I could've used the "and" operand to compare the weight to the specific range of 100-500 newtons.

**Problem 2:**

A software company sells a package that retails for $99. Quantity discounts are given
according to the following table:

|Quantity| Discount|
| --- | --- |
|10–19 |10% |
|20–49 | 20% |
|50–99 | 30% |
|100 or more | 40% |

Write a program that asks the user to enter the number of packages purchased. The program
should then display the amount of the discount (if any) and the total amount of the
purchase after the discount.

__Grades:__ 18 pts for the code (proper commenting 3 pts, reading the inputs correctly 4 pts, output calculation 7 pts, displaying the results in the correct format 4 pts), and 2 pts for reflections (Total 20 pts).

```python
# Stores Package and Discount amount as constants
PACKAGES_PRICE = float(99)
DISCOUNT_1 = .10 # Quantity 10-19
DISCOUNT_2 = .20 # Quantity 20-49
DISCOUNT_3 = .30 # Quantity 50-99
DISCOUNT_4 = .40 # Quantity 100 or more
# Asks user to enter the number of packages purchased
amount_of_packages = int(input('Enter the number of packages purchased: '))
# Calculated the undiscounted total for the packages purchased
undiscounted_total = PACKAGES_PRICE * amount_of_packages
# Compares package amount to determine discount
if amount_of_packages >= 100:
    discount_amount = undiscounted_total * DISCOUNT_4 # Adds 40% discount
    print("You've earned a discount of 40% today!")
    print("Your total today is: $", format(undiscounted_total-discount_amount, '.2f'))
elif amount_of_packages < 100 and amount_of_packages >= 50:
    discount_amount = undiscounted_total * DISCOUNT_3 # Adds 30% discount
    print("You've earned a discount of 30% today!")
    print("Your total today is: $", format(undiscounted_total-discount_amount, '.2f'))
elif amount_of_packages < 50 and amount_of_packages >= 20:
    discount_amount = undiscounted_total * DISCOUNT_2 # Adds 20% discount
    print("You've earned a discount of 20% today!")
    print("Your total today is: $", format(undiscounted_total-discount_amount, '.2f'))
elif amount_of_packages < 20 and amount_of_packages >= 10:
    discount_amount = undiscounted_total * DISCOUNT_1 # Adds 10% discount
    print("You've earned a discount of 10% today!")
    print("Your total today is: $", format(undiscounted_total-discount_amount, '.2f'))
else:
    print("Your total today is: $", format(undiscounted_total, '.2f'))
```

**Problem 4:**
Write a simple robo-advisor program for stock trading. The robo-advisor needs to receive the following inputs from the user
- a cost-basis (e.g. the amount you paid per share for purchasing a stock, here denoated by $C$ in the equations),
- and the current price per share of the corresponding stock (denoated by $P$ in the equations)

The robo-advisor should __use the current price__ and show a message advising the user to exit the trade when the current loss is more than 5\% (i.e., when $P \le C(1-\frac{5}{100})$) or when the current gain is more than 5\% (i.e., $P \ge C(1+\frac{5}{100}$)). __For example__, if a person buys the stocks of XYZ company at \$100.00 per share 

- the robo-advisor should suggest selling (exiting the trade for cutting losses) when the current price is at or below \$95.00 before it gets too late.
- the robo-advisor should suggest selling (exiting the trade for taking profit) when the current price is at or above \$105.00.
- in other cases, robo-advissor should suggest holding the stock.

You may use this example format for displaying the advice to the user: "Currently your gain is \$6.00 (\%6.0) per share and it is advised to take profit and exit the trade" (you can revise accordingly for selling at a loss or staying the in the trade and test it with different inputs). 


__Grades:__ 27 pts for the code (proper comments 5 pts, reading the inputs correctly 6 pts, output calculation 10 pts, displaying the results in the correct format 6 pts), and 3 pts for reflections (Total 30 pts).


```python
# Takes cost-basis input
cost_basis = float(input('Enter the cost-basis: '))

# Take current price per share of the stock as input
pps_input = float(input('Enter the price per share for the corresponding stock: '))

# Sets a gain/loss threshold 
gain_threshold = cost_basis*(1+5/100)
loss_threshold= cost_basis*(1-5/100)

# Calculates the gain/loss relative to current price per share of stock
gain_or_loss = pps_input - cost_basis
gain_loss_percentage = (gain_or_loss/pps_input)*100

# Compares current price per share of stock to gain/loss threshold, then displays
if pps_input <= loss_threshold:
    print(f"Currently your loss is $ {format(gain_or_loss, '.2f')} ({format(gain_loss_percentage, '.1f')} %) per share" )
    print("It is advised to cut losses and exit the trade")
elif pps_input >= gain_threshold:
    print(f"Currently your gain is $ {format(gain_or_loss, '.2f')} ({format(gain_loss_percentage, '.1f')} %) per share" )
    print("It is advised to take profit and exit the trade")
else:
    print(f"Currently your gain is $ {format(gain_or_loss, '.2f')} ({format(gain_loss_percentage, '.1f')} %) per share" )
    print("It is advised to hold your stock")
```

Rather than printing a regular string, I printed an F-string to give me an easier time with formatting the code. Additionally, I added two extra formulas, "gain_or_loss" and "gain_loss_percentage" to show users what percentage their stock has gained or lost.

**Problem 5:**
Write a program that evaluates the score in a shooting game. The score distribution is similar to the following figure (anything outside the target means a zero score).


Here a bullet/arrow can land on any pixel in the 2D world shown by the X-Y coordinates in the image. Your program should 
- receive the x and y coordinates of the bullet/arrow as input (as integer numbers), 
- determine where it is compared to the target, 
- calculate the score, 
- and finally show the score on screen. 

Let's assume that the radius of the circles (from the smallest to the largest) are respectively 5, 10, 15, 20 and 25 pixels and the center of circle is at $x_c$ = 50 pixles and $y_c$ = 50 pixles. Finally, if a test point ends up exactly on the border of two circles, you can consider it as it is a part of circle with the higher score. 



__Hint__: You can use this equation $d = \sqrt{(x - x_c)^2 + (y-y_c)^2}$ to calculate the distance from the center of the target and compare it to the radius of the circles via conditional statements. You can use the exponentiation operator to calculate the square root (i.e., \**0.5).

Some test cases that can help you debug your code:

- for x = 50 and y = 50, or x = 50 and y = 55,  or x = 55 and y = 50, you should see 10 as the score 
- for x = 50 and y = 56, you should see 8 as the score 
- for x = 50 and y = 61, you should see 6 as the score 
- for x = 50 and y = 66, you should see 4 as the score 
- for x = 50 and y = 71, you should see 2 as the score 
- for x = 50 and y = 76, you should see 0 as the score
- you may check other conditions as well

__Grades:__ 24 pts for the code (proper commenting 6 pts, reading the inputs correctly 6 pts, output calculation 9 pts, displaying the results in the correct format 6 pts), and 3 pts for reflections (Total 30 pts).

```python
# Imports math module to get square root function
import math 

# Sets the center of the circle in a list
CENTER = [50, 50]

# Receives x and y coordinate input from user
x_coord = int(input("Enter the x-coordinate: "))
y_coord = int(input("Enter the y-coordinate: "))

# Calculate the distance using the formula
distance_formula = math.sqrt(((x_coord - CENTER[0])**2) + ((y_coord - CENTER[1])**2))

# Determine the score based on distance
if distance_formula <= 5:
    user_score = 10
elif distance_formula <=10:
    user_score = 8
elif distance_formula <=15:
    user_score = 6
elif distance_formula <=20:
    user_score = 4
elif distance_formula <=25:
    user_score = 2
else:
    user_score = 0

# Prints the users score
print(f"You scored {user_score} points!")
```
