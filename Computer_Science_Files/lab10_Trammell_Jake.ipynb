**Problem 2:**
Write  a  program  that  asks  the  user  for  a  number  in  between  1 and  7 and displays the corresponding day of the week (1=Monday, 2=Tuesday, 3 = Wednesday, 4 = Thursday, 5 = Friday, 6 = Saturday, and 7 = Sunday).  Your  program should use __exception handling methods__ correctly to 

- display an error message if the user enters a number that is outside the range of 1 through 7 (using index error). 
- display an error message if the user enters a non-numerical inputs (using value error)

You __CANNOT__ use if-else conditions like Lab 2 for this problem but you will basically be producing the same outcome via exception handling methods.

__Grades:__ 27 pts for the code (proper comments 6 pts, reading and processing the inputs via exception handling methods correctly 21 pts), and 3 pts for reflections (Total 30 pts).


```python
def main():
    # Complete your code here
    # try to execute code
    try:
        # declare a list including the days of week
        days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        # read a number from the user
        day_no = int(input('Enter a number between 1 and 7(represents a day of the week)'))
        # raise Index error if day_no is negative
        if day_no <= 0:
            raise IndexError
        # acces the list 
        print(days_of_week[day_no-1])
    # handle value-related errors
    except ValueError:
        print('ValueError: Please enter a valid number.')
    # handle index related errors
    except IndexError:
        print('IndexError: Please enter a number between 1 and 7.')
main()
```

**Problem 3:**
Complete the sum_of_inverses function that calculates the summation of the inverse values of the enteries in a list of numbers. Your program must handle all of the exceptions that occurs when processing the three lists in the following cell with proper messages.

__Grades:__ 27 pts for the code (proper comments 6 pts, reading and processing the inputs via exception handling methods correctly 21 pts), and 3 pts for reflections (Total 30 pts).

```python
def main():
    # DO NOT CHANGE THE CONTENT OF MAIN FUNCTION!
    list_1 = [1, 2, 3, 4, 5]
    list_2 = [1, '2', 3, 4, 5]
    list_3 = [1, 2, 3, 0, 5]
    print('Testing list 1')
    sum_of_inverses(list_1)
    print('Testing list 2')
    sum_of_inverses(list_2)
    print('Testing list 3')
    sum_of_inverses(list_3)
    
def sum_of_inverses(input_list):   
    # Complete this function.
    # You can simply display the results.
    # No need to return any values
    # Set a running total of inverses
    total = 0
    # create a for loop to run through each item in list
    for item in input_list:
        # try to compute inverse
        try:
            # calculate the the inverse
            inverse = 1/item
            total += inverse
        # raise error if divided by zero
        except ZeroDivisionError:
            print("ZeroDivisionError: Cannot divide by zero")
        # handle TypeErrors
        except TypeError:
            print("TypeError: All elements must be a number")
        # handle ValueErrors
        except ValueError:
            print("ValueError: Invalid number.")

    # Display the final result
    print("Sum of inverses:", total)
    
# Call the main function.
main()
```

**Problem 4:**
Complete the following program so that it 

- asks the user for the last name of one of the UNCW chancellors (current or former)
- checks if the name is not already in the chancellor_names list and manually raises an exception (you can use value error) which shows this message: 'This person has never been a UNCW chancellor!'.
- otherwise, using the chancellor_years list, it shows the time period during which a specific chancellor led UNCW.

__Hint__ You can check if the name is in the list with the 'in' operator. You can effectively find the matching years for each name via the functions you learned for processing the lists! You need to find out the index for the location of the chancellor's name and use it with the corresponding chancellor_years list.



__Grades:__ 27 pts for the code (proper comments 6 pts, reading and processing the inputs via exception handling methods correctly 21 pts), and 3 pts for reflections (Total 30 pts).

```python
def main():
    # DO NOT CHANGE THE FOLLOWING LISTS!
    chancellor_names = ['Hamilton', 'Hoggard', 'Randall', 'Wagoner', 'Leutze', 'DePaolo', 'Miller', 'Sederburg', 'Sartarelli', 'Volety']
    chancellor_years = ['1947-49','1949-58', '1958-68',  '1968-90', '1990-2003', '2003-11', '2011-14', '2014-15', '2015-2022', '2022-pres.']
    # complete your code here
    # ask the user for chancellor's last name
    user_input = input("Enter the last name of a UNCW chancellor") 

    # try to check if the name entered is in the list
    try:
        # Raise a ValueError if name entered is not found on the list
        if user_input not in chancellor_names:
            raise ValueError('This person has never been a UNCW chancellor!')
        # Get chancellors index and print the year they were active
        index = chancellor_names.index(user_input) # finds chancellors name in index
        print(user_input, 'served as chancellor during',chancellor_years[index])
    # handles error if name is not on the list.
    except ValueError as error:
        print(error)

main()
```
