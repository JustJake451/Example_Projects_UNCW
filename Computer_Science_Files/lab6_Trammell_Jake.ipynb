**Problem 2:**
Write a function that asks the users for 10 numbers and __returns__ the average.

__Grades:__ 18 pts for the code (proper comments 4 pts, function definition 11 pts, calling the function correctly 3pts), and 2 pts for reflections (Total 20 pts).


```python
# define the function 
def calc_list_avg():
    # define the amount of inputs
    nums_inputed = 10
    total = 0

    # establishes a for loop to calculate the sum.
    for number in range(nums_inputed):
        number = int(input('Enter a number')) # receives input
        total += number # calculates the sum
    
    # return the average
    average = total/nums_inputed
    return average
    
# call the function
print('The average of the ten numbers is: ',calc_list_avg())
```
