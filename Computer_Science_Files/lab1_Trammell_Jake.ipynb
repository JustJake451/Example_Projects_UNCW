Write a program that converts temperature from F to C. You can use the following equation: 
<br>
$ C = (F-32)\frac{5}{9}$

```python
# Recieves input of temperature in  Fahrenheit and converts it to Celsius 
F = int(input("Insert temperature in Fahrenheit:"))
C = float((F-32)*5/9) # Changing the data type of celsius to a float
print(format(C,'.2f'), "degrees Celsius")
```

**Problem 5:**
Write a program that asks the user for the number of males and the number of females registered in a class. The program should display the percentage of males and females in the class.

__Hint:__  Suppose  there  are  8  males  and  12  females  in  a  class. There  are  20  students  in  the class. The percentage of males can be calculated as $8 \div 20= 0.4$, or 40\%. The percentage of females can be calculated as $12 \div 20 = 0.6$, or 60\%. Show the output with only two decimal points.

__Grades:__ 18 pts for the code (comments 4 pts, reading the inputs correctly 4 pts, output calculation 6 pts, displaying the results 4 pts), and 2 pts for reflections (Total 20 pts).

```python
# Displays percentage of males and females in the class

# Asks for number of males
males = int(input("How many males are in the class?"))

# Asks for number of femals
females = int(input("How many females are in the class?"))

# Asks for number of overall students
total_students = males + females

# Performs percentage calculation for male students
male_percentage = (males/total_students)*100

# Performs percentage calculation for female students
female_percentage = (females/total_students)*100

# Prints the percentages
print('Percentage of male students:', format(male_percentage, '.2f'), '\b%')
print('Percentage of female students:', format(female_percentage, '.2f'), '\b%')
```

**Problem 6:** Write a program that predicts the outcome of your investment based on an expected annual growth rate and investment duration. You can use the following simplified equation: 
<br>
$ w = i(1+g)^d$
<br>
<br>
where $w$ is your expected wealth, $i$ is the initial investment, $g$ is the normalized growth rate (e.g., 50% growth rate means $g = 0.5$), and $d$ is the duration of investment in years. All of these values should be read from a user input.

The output should be shown in this format \$123,456.78. 

Check the results of a \$100.00 investment with a growth rate of 20% in 30 years as a quick test case.

__Grades:__ 31.5 pts for the code (comments 7 pts, reading the inputs correctly 7 pts, output calculation 10.5 pts, displaying the results in the correct format 7 pts), and 3.5 pts for reflections (Total 35 pts).

```python
# Predicts outcome of investment based on expected annual growth rate and duration.

# Creating inputable variables for the formula
initial_amount = float(input("Enter initial investment amount: "))
growth_rate = float(input("Enter the growth rate as a decimal: "))
duration = int(input("Enter the duration of the investment: "))

# Performs calculation of expected wealth
expected_wealth = float(initial_amount * (1+growth_rate)**duration)

# Prints for user to read
print("Expected wealth: $",format(expected_wealth, ',.2f'))
```
